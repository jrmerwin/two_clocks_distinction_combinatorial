# Generate graph-index figures

Run from the repository root or from the notebooks directory.

```
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

cwd = Path.cwd()
if cwd.name == 'notebooks':
    root = cwd.parent
elif (cwd/'data').exists() and (cwd/'figures').exists():
    root = cwd
else:
    root = Path('..').resolve()

data = root/'data'
figdir = root/'figures'
figdir.mkdir(exist_ok=True)
print(root)
```

```
s = pd.read_csv(data/'stage_i_exact41over40_summary.csv')
fig, ax = plt.subplots(figsize=(7,4.3))
ax.bar(s['branch'], s['mean_highl_chi2'])
ax.set_ylabel('Mean high-ell TTTEEE_lite chi-square')
ax.set_title('Exact 41/40 branch vs graph and continuum branches')
ax.tick_params(axis='x', rotation=20)
fig.tight_layout()
fig.savefig(figdir/'graph_index_highl_chi2.pdf')
fig.savefig(figdir/'graph_index_highl_chi2.png', dpi=200)
```

```
g = pd.read_csv(data/'stage_g_broad_A_profile.csv')
fig, ax = plt.subplots(figsize=(7,4.3))
x = range(len(g))
labels = [f"{r.likelihood}\n{r.objective}" for r in g.itertuples()]
width=0.35
ax.bar([i-width/2 for i in x], g['deu_A_planck'], width, label='DEU')
ax.bar([i+width/2 for i in x], g['native_A_planck'], width, label='Native')
ax.axhline(41/40, linestyle='--', linewidth=1, label='41/40')
ax.axhline(1.0, linestyle=':', linewidth=1, label='A=1')
ax.set_xticks(list(x), labels, rotation=20, ha='right')
ax.set_ylabel('Profiled A_planck')
ax.set_title('Broad-amplitude profile separates graph and continuum basins')
ax.legend()
fig.tight_layout()
fig.savefig(figdir/'broad_A_basins.pdf')
fig.savefig(figdir/'broad_A_basins.png', dpi=200)
```

```
e = pd.read_csv(data/'stage_e_full_plik_profile.csv')
fig, ax = plt.subplots(figsize=(7,4.3))
labels = [f"{r.point}\n{r.objective.replace('_',' ')}" for r in e.itertuples()]
ax.bar(labels, e['native_minus_deu_highl_chi2'])
ax.axhline(0, linewidth=1)
ax.set_ylabel('Native minus DEU full high-ell chi-square')
ax.set_title('Full-Plik nuisance-profile stress test')
ax.tick_params(axis='x', rotation=20)
fig.tight_layout()
fig.savefig(figdir/'full_plik_profile_highl_gaps.pdf')
fig.savefig(figdir/'full_plik_profile_highl_gaps.png', dpi=200)
```

```
f = pd.read_csv(data/'stage_f_fixed_unity_profile.csv')
fig, ax = plt.subplots(figsize=(7,4.3))
labels = [f"{r.likelihood}\n{r.objective.replace('_',' ')}" for r in f.itertuples()]
ax.bar(labels, f['native_minus_deu_highl_chi2'])
ax.axhline(0, linewidth=1)
ax.set_ylabel('Native minus DEU high-ell chi-square')
ax.set_title('Fixed A_planck=1 ablation: DEU fails')
ax.tick_params(axis='x', rotation=20)
fig.tight_layout()
fig.savefig(figdir/'fixed_unity_failure.pdf')
fig.savefig(figdir/'fixed_unity_failure.png', dpi=200)
```
